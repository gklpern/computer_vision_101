```python
import cv2
import numpy as np
```

```python
image = cv2.imread("Dhaulagiri_mountain.jpg")
```

```python
cv2.imshow("Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
import matplotlib.pyplot as plt
plt.imshow(image_rgb)
plt.show()
```

## Resizing


```python
print(image_rgb.shape)
```

```python
new_height = 300
new_width = 300

resized_image = cv2.resize(image_rgb, (new_width, new_height))
```

```python
plt.imshow(resized_image)
plt.show()
```

## Image Scaling

```python
scale_percent = 50 
```

```python
s_width = int(image_rgb.shape[1] * scale_percent / 100)
s_height = int(image_rgb.shape[1] * scale_percent / 100)
```

```python
scaled_image = cv2.resize(image_rgb, (s_width, s_height), interpolation = cv2.INTER_AREA)
```

```python
plt.imshow(scaled_image)
```
